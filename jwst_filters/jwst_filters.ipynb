```python
import matplotlib.pyplot as plt
from species import SpeciesInit
from species.read.read_filter import ReadFilter

SpeciesInit()
```

## Filters

Pulling them from https://svo2.cab.inta-csic.es/theory/fps/index.php?mode=browse&gname=JWST&asttype=


```python
f10 = ReadFilter('JWST/MIRI.F1065C')
f11 = ReadFilter('JWST/MIRI.F1140C')
f15 = ReadFilter('JWST/MIRI.F1550C')

f10_filter_array = f10.get_filter()
f10_wavelengths = f10_filter_array[:, 0]
f10_transmission = f10_filter_array[:, 1]

f11_filter_array = f11.get_filter()
f11_wavelengths = f11_filter_array[:, 0]
f11_transmission = f11_filter_array[:, 1]

f15_filter_array = f15.get_filter()
f15_wavelengths = f15_filter_array[:, 0]
f15_transmission = f15_filter_array[:, 1]
fig, ax = plt.subplots()
ax.plot(f10_wavelengths, f10_transmission, label='F1065C')
ax.plot(f11_wavelengths, f11_transmission, label='F1140C')
ax.plot(f15_wavelengths, f15_transmission, label='F1550C')
ax.set_xlabel('Wavelength (µm)')
ax.set_ylabel('Transmission')
ax.set_title('JWST MIRI Filter Transmission')
ax.legend()
fig.savefig('jwst_miri_filters.png', dpi=300)
plt.show()
```
