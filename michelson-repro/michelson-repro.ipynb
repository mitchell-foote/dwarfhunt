```python
import numpy as np
from species import SpeciesInit
from species.data.database import Database
from species.read.read_model import ReadModel
from species.read.read_filter import ReadFilter

SpeciesInit()

db = Database()

db.add_model('sonora-elfowl-t', wavel_range=(.61, 18.), wavel_sampling=1000.0)
db.add_model('sonora-elfowl-y', wavel_range=(.61, 18.), wavel_sampling=1000.0)
db.add_filter('JWST/MIRI.F1065C')
db.add_filter('JWST/MIRI.F1140C')
db.add_filter('JWST/MIRI.F1550C')

model = ReadModel('sonora-elfowl-t', wavel_range=(.61, 18.))
model_y = ReadModel('sonora-elfowl-y', wavel_range=(.61, 18.))




model_bounds = model.get_bounds()
# loop through the keys in the dictionary to get the tuple values, then generate 200 random values based between the min and max of each tuple, and store them in an array
model_random_values = {key: np.random.uniform(low=val[0], high=val[1], size=200) for key, val in model_bounds.items()}
model_random_values['radius'] = np.random.uniform(low=0.6, high=1.3, size=200)
model_random_values['distance'] = np.full(200, 10.0)

model_y_bounds = model_y.get_bounds()
model_y_random_values = {key: np.random.uniform(low=val[0], high=val[1], size=200) for key, val in model_y_bounds.items()}
common_keys = set(model_random_values.keys()) & set(model_y_random_values.keys())
full_planet_array_1 = np.column_stack([model_random_values[key] for key in common_keys])
full_planet_array_2 = np.column_stack([model_y_random_values[key] for key in common_keys])

model_t_F1065C = ReadModel('sonora-elfowl-t', wavel_range=(.61, 18.), filter_name="JWST/MIRI.F1065C")
model_t_F1140C = ReadModel('sonora-elfowl-t', wavel_range=(.61, 18.), filter_name="JWST/MIRI.F1140C")
model_t_F1550C = ReadModel('sonora-elfowl-t', wavel_range=(.61, 18.), filter_name="JWST/MIRI.F1550C")

first_model_planet_1 = {key: model_random_values[key][0] for key in model_random_values.keys()}

planet_1_flux1065 = model_t_F1065C.get_flux(first_model_planet_1)
planet_1_flux1140 = model_t_F1140C.get_flux(first_model_planet_1)
planet_1_flux1550 = model_t_F1550C.get_flux(first_model_planet_1)

planet_1_mags : tuple[float, float] = model_t_F1065C.get_magnitude(first_model_planet_1) # type: ignore

first_model_planet_1['flux_1065C'] = planet_1_flux1065
first_model_planet_1['flux_1140C'] = planet_1_flux1140
first_model_planet_1['flux_1550C'] = planet_1_flux1550
first_model_planet_1['abs_mag_1065'] = planet_1_mags[0]

print(first_model_planet_1)

# planet_1_box = model.get_flux(first_model_planet_1, filter_list=['JWST/MIRI.F1065C', 'JWST/MIRI.F1140C', 'JWST/MIRI.F1550C'])
# first_model_planet_1['flux'] = planet_1_box.flux
# first_model_planet_1['wavelength'] = planet_1_box.wavelength

```
